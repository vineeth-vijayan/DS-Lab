# 3

```python
f = open("Value", 'r')
x,y = [],[]
for row in f.readlines():
    x.append(int(row[1:-2].split(",")[0]))
    y.append(int(row[1:-2].split(",")[1]))
plt.plot(x,y,'*-g')
plt.title("Plot Title")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
x1,y1,x2,y2 = [],[],[],[]
for i in range(5):
    x1.append(int(np.random.random()*10))
    x2.append(int(np.random.random()*10))
    y1.append(int(np.random.random()*10))
    y2.append(int(np.random.random()*10))
plt.plot(x1,y1,color='red', label='line one')
plt.plot(x2,y2,color='green', label='line two')
plt.legend()
plt.show()
```

```python
x1,y1,x2,y2 = [],[],[],[]
x3,y3,x4,y4 = [],[],[],[]
for i in range(5):
    x1.append(int(np.random.random()*10))
    x2.append(int(np.random.random()*10))
    y1.append(int(np.random.random()*10))
    y2.append(int(np.random.random()*10))
    x3.append(int(np.random.random()*10))
    x4.append(int(np.random.random()*10))
    y3.append(int(np.random.random()*10))
    y4.append(int(np.random.random()*10))
plt.subplot(2,2,1)
plt.plot(x1,y1)
plt.subplot(2,2,2)
plt.plot(x2,y2)
plt.subplot(2,2,3)
plt.plot(x3,y3)
plt.subplot(2,2,4)
plt.plot(x4,y4)
plt.show()
```
